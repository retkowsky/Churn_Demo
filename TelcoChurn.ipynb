```python
# File location and type
file_location = "/FileStore/tables/churn.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
churn = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(churn)
```

```python
# Visualisation données Churn
display(churn)
```

```python
customer=customer.withColumn('Children', customer['Children'].astype("float"))
customer=customer.withColumn('Est Income', customer['Est Income'].astype("float"))
customer=customer.withColumn('Age', customer['Age'].astype("float"))
customer=customer.withColumn('LongDistance', customer['LongDistance'].astype("float"))
customer=customer.withColumn('International', customer['International'].astype("float"))
customer=customer.withColumn('Local', customer['Local'].astype("float"))
customer=customer.withColumn('Dropped', customer['Dropped'].astype("float"))
customer=customer.withColumn('Usage', customer['Usage'].astype("float"))
customer.dtypes
```

```python
customer = customer.withColumnRenamed("ID", "ID2")
```

```python
display(customer)
```

<a id="etape2"></a>
## Etape 2 : Préparation des données (jointure Spark SQL)

```python
customer.createOrReplaceTempView("customer")
churn.createOrReplaceTempView("churn")

requeteSQL = """

SELECT c.*
    , p.ID
    , p.CHURN 
FROM customer c
INNER JOIN churn p
    on c.ID2 = p.ID
    
"""
merged = spark.sql(requeteSQL)
display(merged)
```

<a id="etape3"></a>
## Etape 3 : Transformation des données

```python
merged = merged.withColumnRenamed("Est Income", "EstIncome").withColumnRenamed("Car Owner","CarOwner")
display(merged)
```

## Etape 5 : Construction du pipeline SPARK - Modélisation Random Forest 
https://spark.apache.org/docs/2.0.2/ml-guide.html

```python
from pyspark.ml.feature import StringIndexer, VectorIndexer
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import RandomForestClassifier

# String indexers
stringIndexer1 = StringIndexer(inputCol='Gender', outputCol='GenderEncoded')
stringIndexer2 = StringIndexer(inputCol='Status',outputCol='StatusEncoded')
stringIndexer3 = StringIndexer(inputCol='CarOwner',outputCol='CarOwnerEncoded')
stringIndexer4 = StringIndexer(inputCol='Paymethod',outputCol='PaymethodEncoded')
stringIndexer5 = StringIndexer(inputCol='LocalBilltype',outputCol='LocalBilltypeEncoded')
stringIndexer6 = StringIndexer(inputCol='LongDistanceBilltype',outputCol='LongDistanceBilltypeEncoded')
stringIndexer7 = StringIndexer(inputCol='RatePlan',outputCol='RatePlanEncoded')
stringIndexer8 = StringIndexer(inputCol='CHURN', outputCol='label')

# Pipeline API
assembler = VectorAssembler(inputCols=["GenderEncoded", "StatusEncoded", "CarOwnerEncoded", "PaymethodEncoded", "LocalBilltypeEncoded", \
                                       "LongDistanceBilltypeEncoded", "Children", "EstIncome", "Age", "LongDistance", "International", "Local",\
                                      "Dropped","Usage","RatePlanEncoded"], outputCol="features")


# Instantiation de l'algorithme
rf=RandomForestClassifier(labelCol="label", featuresCol="features")

#pipeline = Pipeline(stages=[stringIndexer1, stringIndexer2, stringIndexer3, assembler, rf])
pipeline = Pipeline(stages=[stringIndexer1,stringIndexer2,stringIndexer3,stringIndexer4,stringIndexer5,stringIndexer6,stringIndexer7, stringIndexer8, assembler, rf])
```

```python
# Partitionnement des données
train, test = merged.randomSplit([0.8,0.2], seed=6)
```

```python
apprentissage=train.toPandas()
print ('Lignes et colonnes base Apprentissage :')
print(apprentissage.shape)
```

```python
testbase=test.toPandas()
print("Lignes et colonnes base de Test :")
print(testbase.shape)
```

```python
import time
debutmodele=time.time()

# Estimation du modèle
model = pipeline.fit(train)

# Temps de construction du modèle
finmodele=time.time()
print("Temps d'estimation du modèle = " + str(finmodele-debutmodele))+ " secondes"
```

<a id="etape6"></a>
## Etape 6 : Validation du modèle - Scoring sur la base de Test

```python
results = model.transform(test)
display(results)
```

## Etape 7 : Validation du modèle - Indicateurs de qualité du modèle

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluator = BinaryClassificationEvaluator(rawPredictionCol="prediction", labelCol="label", metricName="areaUnderROC")
print 'Aire sous la courbe (AUC) = {:.2f}.'.format(evaluator.evaluate(results))
```
